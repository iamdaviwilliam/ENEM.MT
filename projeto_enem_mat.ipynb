```python
df_2018 = pd.read_csv('ITENS_PROVA_2018.csv', sep=";", encoding="ISO-8859-1")

chunks = pd.read_csv('MICRODADOS_ENEM_2018.csv', sep=";", encoding="ISO-8859-1", chunksize=500)

md_2018_chunks = []

for chunk in chunks:
    md_2018_chunks.append(chunk)

md_2018 = pd.concat(md_2018_chunks, ignore_index=True)
df_concatenado = pd.concat([df_2018, md_2018], axis=0, ignore_index=True)

display(df_concatenado)
```

```python
colunas_desejadas = ['TP_FAIXA_ETARIA', 'TP_ENSINO', 'SG_UF_ESC', 'TP_PRESENCA_MT', 'TP_COR_RACA', 
                     'NU_NOTA_MT'] 

df_filtrado = df_concatenado[colunas_desejadas]

display(df_filtrado)
```

```python
# Substitui valores NaN em 'TP_COR_RACA' por -1
df_filtrado.loc[:, 'TP_COR_RACA'] = df_filtrado['TP_COR_RACA'].fillna(-1)

# Substitui 0 por -1 em 'TP_COR_RACA'
df_filtrado.loc[:, 'TP_COR_RACA'] = df_filtrado['TP_COR_RACA'].replace(0, -1)

# Substitui valores 0 em 'NU_NOTA_MT' pela mÃ©dia e preenche NaNs com a mÃ©dia
media_nota_mt = df_filtrado['NU_NOTA_MT'].mean()
df_filtrado.loc[:, 'NU_NOTA_MT'] = df_filtrado['NU_NOTA_MT'].replace(0, media_nota_mt).fillna(media_nota_mt)

# Substitui valores NaN em 'TP_FAIXA_ETARIA' pela mediana
df_filtrado.loc[:, 'TP_FAIXA_ETARIA'] = df_filtrado['TP_FAIXA_ETARIA'].fillna(df_filtrado['TP_FAIXA_ETARIA'].median())

# Substitui 0 em 'TP_FAIXA_ETARIA' pela mediana
df_filtrado.loc[:, 'TP_FAIXA_ETARIA'] = df_filtrado['TP_FAIXA_ETARIA'].replace(0, df_filtrado['TP_FAIXA_ETARIA'].median())

# Substitui valores NaN em 'TP_ENSINO' com o valor mais frequente e substitui 0 por 3
modo_tp_ensino = df_filtrado['TP_ENSINO'].mode()[0]
df_filtrado.loc[:, 'TP_ENSINO'] = df_filtrado['TP_ENSINO'].replace(0, 3).fillna(modo_tp_ensino)

# Substitui valores NaN em 'TP_PRESENCA_MT' por -1
df_filtrado.loc[:, 'TP_PRESENCA_MT'] = df_filtrado['TP_PRESENCA_MT'].fillna(-1)

# Substitui valores NaN em 'SG_UF_ESC' por 'ND' (NÃ£o DisponÃ­vel)
df_filtrado.loc[:, 'SG_UF_ESC'] = df_filtrado['SG_UF_ESC'].fillna('ND')

# Exibe o DataFrame atualizado
display(df_filtrado)
```

```python
media_estado = df_filtrado.groupby('SG_UF_ESC')['NU_NOTA_MT'].mean()

media_tipo_ensino = df_filtrado.groupby('TP_ENSINO')['NU_NOTA_MT'].mean()

print("MÃ©dia das notas por estado:")
print(media_estado)

print("\nMÃ©dia das notas por tipo de instituiÃ§Ã£o (1 = PÃºblica, 2 = Particular):")
print(media_tipo_ensino)
```

```python
mapeamento_estado_regiao = {
    'AC': 'Norte', 'AP': 'Norte', 'AM': 'Norte', 'PA': 'Norte', 'RO': 'Norte', 'RR': 'Norte', 'TO': 'Norte',
    'AL': 'Nordeste', 'BA': 'Nordeste', 'CE': 'Nordeste', 'MA': 'Nordeste', 'PB': 'Nordeste',
    'PE': 'Nordeste', 'PI': 'Nordeste', 'RN': 'Nordeste', 'SE': 'Nordeste',
    'ES': 'Sudeste', 'MG': 'Sudeste', 'RJ': 'Sudeste', 'SP': 'Sudeste',
    'PR': 'Sul', 'RS': 'Sul', 'SC': 'Sul',
    'DF': 'Centro-Oeste', 'GO': 'Centro-Oeste', 'MT': 'Centro-Oeste', 'MS': 'Centro-Oeste'
}

df_filtrado['REGIAO'] = df_filtrado['SG_UF_ESC'].map(mapeamento_estado_regiao)

media_regiao = df_filtrado.groupby('REGIAO')['NU_NOTA_MT'].mean()

print("MÃ©dia das notas por regiÃ£o:")
print(media_regiao)
```

```python
media_notas_mt_por_raca = df_filtrado.groupby('TP_COR_RACA')['NU_NOTA_MT'].mean()

print("MÃ©dia das notas de MatemÃ¡tica por raÃ§a:")
print(media_notas_mt_por_raca)
```

```python
media_por_estado_tipo_escola = df_filtrado.groupby(['SG_UF_ESC', 'TP_ENSINO'])['NU_NOTA_MT'].mean().unstack()

print("MÃ©dia das notas de matemÃ¡tica por estado e tipo de escola:")
print("1 = PÃºblica, 2 = Particular")
print(media_por_estado_tipo_escola)
```
